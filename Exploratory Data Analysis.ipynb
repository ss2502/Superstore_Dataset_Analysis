```python

# extract rows with Category 'Appliances'
appliances_df = data.loc[data['Category'] == 'Appliances']

# calculate total sales value for Appliances category
total_sales = appliances_df['Sales'].sum()

print('Total sales value for Appliances category:', total_sales)
```

```python
#In class assignment 

# Create a dataframe with columns 'State' and 'Sales'
state_sales_df = data[['State', 'Sales']]

# Group by state and calculate the average sales
state_avg_sales = state_sales_df.groupby('State').mean()

# Sort the dataframe by average sales in descending order
sorted_avg_sales = state_avg_sales.sort_values(by='Sales', ascending=False)

# Find the highest average sales
highest_avg_sales = round(sorted_avg_sales.iloc[0]['Sales'])

print("The highest average sales in a state is:", highest_avg_sales)
```

```python
state_avg_sales
```

```python

# Create a new column for total cost (Sales + Shipping Cost)
data['Total Cost'] = data['Sales'] + data['Shipping Cost']

# Sort the dataframe by total profit by category in descending order
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index().sort_values('Profit', ascending=False)

# Find the highest total profit for a category and round to nearest whole number
highest_profit = round(profit_by_category['Profit'].max())

print('The highest total profit for a category is:', highest_profit)
```

```python


# Extract the required columns and calculate total sales
sales_df = data[['Category', 'Region', 'Sales']]
sales_df['Total Sales'] = sales_df['Sales']

# Group by category and region and sum the sales
sales_grouped_df = sales_df.groupby(['Category', 'Region']).agg({'Total Sales': 'sum'})

# Sort by total sales in descending order
sorted_sales_df = sales_grouped_df.sort_values('Total Sales', ascending=False)

# Find the highest total sales for a combination of category and region
highest_sales = int(sorted_sales_df.iloc[0]['Total Sales'].round())
print(highest_sales)
```

```python
# Plot a histogram of the sales column
plt.hist(data['Sales'])
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot a scatter plot of sales vs. profit
sns.scatterplot(x='Sales', y='Profit', data=data)
plt.title('Scatter Plot of Sales vs. Profit')
plt.show()
```

```python
# Plot a bar chart of the total sales by category
data.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Create a line plot
data.plot.line(x='Ship_Date', y='Sales')

# Add labels and title
plt.xlabel('Shipping Date')
plt.ylabel('Sales')
plt.title('Sales variation')

# Show the plot
plt.show()
```

```python
# Check for missing data
print(data.isnull().sum())
```

```python
# Check for duplicate values
print(data.duplicated().sum())
```

```python
# Calculate the Z-score for each value in the Sales column
z_scores = np.abs((data['Sales'] - data['Sales'].mean()) / data['Sales'].std())

# Find the outliers (values with Z-score > 3)
outliers = data[z_scores > 3]

# Print the outliers
print(outliers)
```

```python
# Create a box plot of the sales column
plt.boxplot(data['Sales'])
plt.show()
```

```python
# Remove outliers with a z-score greater than 3
data = data[z_scores < 3]
```

```python
data.head
```

```python
# Group the data by region and category and calculate the total profit for each group
profit_by_region_category = data.groupby(['Region', 'Category'])['Profit'].sum()

# Pivot the data to create a table with regions as rows, categories as columns, and profit as values
profit_table = profit_by_region_category.reset_index().pivot(index='Category', columns='Region', values='Profit')

# Create a stacked bar chart of profit by region and category
profit_table.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and titles
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Category and Region')

# Show the plot
plt.show()
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot the sales by category in the first subplot
sales_by_category = data.groupby('Category')['Sales'].sum()
sales_by_category.plot(kind='bar', ax=ax1)
ax1.set_title('Sales by Category')

# Plot the sales by region in the second subplot
sales_by_region = data.groupby('Region')['Sales'].sum()
sales_by_region.plot(kind='bar', ax=ax2)
ax2.set_title('Sales by Region')

# Display the subplots
plt.show()
```
